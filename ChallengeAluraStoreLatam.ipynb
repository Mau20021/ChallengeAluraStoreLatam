### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```


# INICIO DE DESAFIO

***Introducción***

El presente informe tiene como objetivo analizar el desempeño de las cuatro tiendas de la cadena Alura Store con el fin de determinar cuál de ellas debería vender el Sr. Juan para financiar un nuevo emprendimiento. Para ello, se evaluaron diversos indicadores clave, incluyendo los ingresos totales, las categorías de productos más y menos vendidas, las calificaciones promedio de los clientes, los productos con mayor y menor facturación y el costo promedio de envío por tienda.

El análisis se basó en datos históricos de ventas y en visualizaciones que permiten comparar de manera clara el rendimiento de cada tienda.

```python
import matplotlib.pyplot as plt

tienda['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

df = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)
df.head()
```

#1. Análisis de facturación



```python
facturacion = df.groupby('Tienda')['Precio'].sum().sort_values(ascending=False)

facturacion.plot(kind='bar')
plt.title('Facturación total por tienda')
plt.ylabel('Facturación')
plt.show()

facturacion
```

**Ingresos totales**

La facturación total muestra diferencias claras entre las tiendas. La Tienda 1 lidera en ingresos, seguida por las Tiendas 2 y 3. En contraste, la Tienda 4 presenta la menor facturación global, lo que indica un menor volumen de ventas y una contribución más baja a los ingresos de la cadena.

# 2. Ventas por categoría

```python
vc1 = tienda.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)

vc1.plot(kind='bar')
plt.title('Ventas por categoria en tienda 1')
plt.ylabel('Facturación')
plt.show()

vc1
```

```python
vc2 = tienda2.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)

vc2.plot(kind='bar')
plt.title('Ventas por categoria en tienda 2')
plt.ylabel('Facturación')
plt.show()

vc2
```

```python
vc3 = tienda3.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)

vc3.plot(kind='bar')
plt.title('Ventas por categoria en tienda 3')
plt.ylabel('Facturación')
plt.show()

vc3
```

```python
vc4 = tienda4.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)

vc4.plot(kind='bar')
plt.title('Ventas por categoria en tienda 4')
plt.ylabel('Facturación')
plt.show()

vc4
```

**Categorías de productos más y menos vendidas**

En todas las tiendas, las categorías que generan mayores ingresos son Electrónicos, Electrodomésticos y Muebles, lo que sugiere que los productos de alto valor económico son el principal motor de ventas.

Sin embargo, la Tienda 4 muestra ingresos considerablemente menores en estas categorías clave, especialmente en Electrodomésticos, lo que evidencia un menor desempeño comercial frente a las demás tiendas. Las categorías de menor venta, como Libros y Artículos para el hogar, tienen un impacto reducido en todas las tiendas.

# 3. Calificación promedio de la tienda


```python
calificacionprom = df.groupby('Tienda')['Calificación'].mean().sort_values(ascending=False)

calificacionprom.plot(kind='bar')
plt.title('Calificación promedio por tienda')
plt.ylabel('Calificacion promedio')
plt.show()

calificacionprom
```

**Calificaciones promedio de los clientes**

Las evaluaciones de los clientes reflejan niveles de satisfacción similares entre las tiendas, con valores cercanos a 4 sobre 5. La Tienda 3 presenta la mejor calificación promedio, mientras que la Tienda 1 tiene la más baja. La Tienda 4 se sitúa en una posición intermedia, sin destacar ni positiva ni negativamente en este aspecto.

Esto indica que la experiencia del cliente en la Tienda 4 es aceptable, pero no constituye una ventaja competitiva significativa.

# 4. Productos más y menos vendidos

```python
productos1 = tienda.groupby('Producto')['Precio'].sum().sort_values(ascending=False)

# Top 5 y Bottom 5
top5_1 = productos1.head(5)
bottom5_1 = productos1.tail(5)

# Unir ambos
productos_1 = pd.concat([top5_1, bottom5_1])

productos_1.plot(kind='bar')
plt.title('Productos más y menos vendidos en tienda 1')
plt.ylabel('Facturación')
plt.show()

productos_1
```

```python
productos2 = tienda2.groupby('Producto')['Precio'].sum().sort_values(ascending=False)

# Top 5 y Bottom 5
top5_2 = productos2.head(5)
bottom5_2 = productos2.tail(5)

# Unir ambos
productos_2 = pd.concat([top5_2, bottom5_2])

productos_2.plot(kind='bar')
plt.title('Productos más y menos vendidos en tienda 2')
plt.ylabel('Facturación')
plt.show()

productos_2
```

```python
productos3 = tienda3.groupby('Producto')['Precio'].sum().sort_values(ascending=False)

# Top 5 y Bottom 5
top5_3 = productos3.head(5)
bottom5_3 = productos3.tail(5)

# Unir ambos
productos_3 = pd.concat([top5_3, bottom5_3])

productos_3.plot(kind='bar')
plt.title('Productos más y menos vendidos en tienda 3')
plt.ylabel('Facturación')
plt.show()

productos_3
```

```python
productos4 = tienda4.groupby('Producto')['Precio'].sum().sort_values(ascending=False)

# Top 5 y Bottom 5
top5_4 = productos4.head(5)
bottom5_4 = productos4.tail(5)

# Unir ambos
productos_4 = pd.concat([top5_4, bottom5_4])

productos_4.plot(kind='bar')
plt.title('Productos más y menos vendidos en tienda 4')
plt.ylabel('Facturación')
plt.show()

productos_4
```

**Productos más y menos vendidos**

Los productos que generan mayor facturación en todas las tiendas corresponden a artículos tecnológicos y electrodomésticos de alto costo, como televisores, teléfonos inteligentes y refrigeradores. Estos productos concentran gran parte de los ingresos.

En la Tienda 4, aunque estos productos también lideran las ventas, los montos obtenidos son inferiores a los de otras tiendas, lo que refuerza la tendencia de menor rendimiento global.

Por otro lado, los productos menos vendidos corresponden a artículos de bajo precio, principalmente juguetes y accesorios, cuyo impacto en los ingresos totales es limitado en todas las tiendas.

# 5. Envío promedio por tienda

```python
df.columns
```

```python
enviosprom = df.groupby('Tienda')['Costo de envío'].mean().sort_values(ascending=True)

enviosprom.plot(kind='bar')
plt.title('Costo promedio de envío por tienda')
plt.ylabel('Costo de envío promedio')
plt.show()

enviosprom
```

**Costo promedio de envío**

La Tienda 4 presenta el costo de envío promedio más bajo entre las cuatro tiendas, lo cual representa una fortaleza logística potencial. Sin embargo, esta ventaja no se traduce en un mayor volumen de ventas ni en ingresos superiores, lo que sugiere que otros factores influyen más en el desempeño comercial.

***ANÁLISIS FINAL***

Tras analizar todos los indicadores considerados, se concluye que la Tienda 4 es la que presenta el desempeño global más bajo dentro de la cadena Alura Store.

Esta conclusión se fundamenta principalmente en que:

- Posee la menor facturación total.

- Presenta menores ingresos en las categorías de productos más importantes.

- No destaca en calificaciones de clientes ni en productos líderes.

- Su ventaja en costos de envío no compensa su bajo volumen de ventas.

Por lo tanto, se recomienda que el Sr. Juan venda la Tienda 4 para financiar su nuevo emprendimiento, ya que su desinversión tendría el menor impacto negativo en el rendimiento general de la empresa y permitiría optimizar los recursos hacia unidades más rentables.
